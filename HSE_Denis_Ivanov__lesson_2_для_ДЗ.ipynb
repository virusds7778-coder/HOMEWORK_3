```python
def factorial(n):
    if n < 0:
        return "Факториал не определен для отрицательных чисел"
    elif n == 0:
        return 1
    else:
        res = 1
        for i in range(1, n + 1):
            res *= i
        return res
```

```python
def find_max_of_three(numbers_tuple):
    if not isinstance(numbers_tuple, tuple) or len(numbers_tuple) != 3:
        return "Ошибка: Аргумент должен быть кортежем из трех чисел"
    try:
        return max(numbers_tuple)
    except TypeError:
        return "Ошибка: Кортеж должен содержать только числа"
```

```python
def calculate_right_triangle_area(cathetus1, cathetus2):
    if not isinstance(cathetus1, (int, float)) or not isinstance(cathetus2, (int, float)):
        return "Ошибка: Катеты должны быть числами"
    if cathetus1 <= 0 or cathetus2 <= 0:
        return "Ошибка: Катеты должны быть положительными числами"
    return 0.5 * cathetus1 * cathetus2
```

```python
import re

# Placeholder for court data (can be loaded from a file or extended)
COURT_DATA = {
    "A40": {
        "name": "Арбитражный суд города Москвы",
        "address": "115225, г. Москва, ул. Б. Тульская, 17"
    }
    # Add more courts here as needed
}

# Plaintiff data (your data)
PLAINTIFF_DATA = {
    "name": "Иванов Иван Иванович",
    "inn": "7701000000",
    "ogrnip": "300770000000000",
    "address": "101000, г. Москва, ул. Ленина, 1"
}

def generate_document_header(input_data):
    case_number = input_data.get("case_number")
    if not case_number:
        return "Ошибка: В аргументе отсутствует 'case_number'."

    # Extract court code from case number (e.g., A40 from A40-123456/2023)
    court_code_match = re.match(r"([A-Z0-9]+)-\d+/", case_number)
    court_code = court_code_match.group(1) if court_code_match else None

    if not court_code or court_code not in COURT_DATA:
        return f"Ошибка: Код суда '{court_code}' из номера дела не найден или номер дела имеет неверный формат."

    court = COURT_DATA[court_code]

    respondent_data = input_data.get("respondent", {})
    respondent_name = respondent_data.get("name", "Не указано")
    respondent_inn = respondent_data.get("inn", "Не указано")
    respondent_ogrn = respondent_data.get("ogrn", "Не указано") # Assuming OGRN for organizations
    respondent_address = respondent_data.get("address", "Не указано")

    header = f"""
В {court['name']}
Адрес: {court['address']}

Истец: {PLAINTIFF_DATA['name']}
ИНН{PLAINTIFF_DATA['inn']} ОГРНИП {PLAINTIFF_DATA['ogrnip']}
Адрес: {PLAINTIFF_DATA['address']}

Ответчик: {respondent_name}
ИНН{respondent_inn} ОГРН {respondent_ogrn}
Адрес: {respondent_address}

Номер дела {case_number}
"""
    return header.strip()
```

```python
def generate_multiple_document_headers(respondents_list):
    print("--- Генерация нескольких шапок документов ---\n")
    for respondent_data in respondents_list:
        # Assuming each item in respondents_list is a dictionary suitable for generate_document_header
        document_header = generate_document_header(respondent_data)
        print(document_header)
        print("\n---\n") # Separator for readability
```
